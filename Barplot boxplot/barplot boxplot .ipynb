```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

# Load Data

```python
data = pd.read_csv('countries.csv')
data.head()
```

```python
data_2007 = data[data.year == 2007]
data_2007.head()
```

```python
top10 = data_2007.sort_values('population', ascending = False).head(10)
```

```python
# plt.bar
x = range(10)
plt.bar(x, top10.population/10**6)
plt.xticks(x, top10.country, rotation = 'vertical')
plt.title("10 most populous countries")
plt.ylabel("population in millions")
plt.show()
```

```python
#sns.barplot
# question: how to operate data in sns
g = sns.barplot(x = "country", y = "population",data = top10)
g.set_title("10 most populous countries")
plt.show()
```

```python
# boxplot
df = data.sort_values('population', ascending = False)
box = sns.boxplot(x = "country", y = "population", hue = "year", data = df)
box.set_title("10 most populous countries")
plt.show()
```

```python
# subplots
fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize = (18,6))
sns.barplot(x = "country", y = "population", data = top10, ax= ax[0])
ax[0].set_title("10 most populous countries")
sns.barplot(x = "country", y = "gdpPerCapita", data = top10, ax =ax[1])
ax[1].set_title("gdpPerCapita of 10 most populous country")
ax[1].set_yscale('log')
plt.tight_layout()
plt.show()
```
