```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

# Load Data

```python
data = pd.read_csv("countries.csv")
data.head()
```

```python
data_2007 = data[data.year == 2007]
data_2007.head()
```

```python
# plt.scatter
plt.scatter(data_2007.gdpPerCapita, data_2007.lifeExpectancy,5)
plt.xlabel("gdp per capita ($)")
plt.ylabel("life expectancy")
plt.show()
```

```python
# sns.swarmplot
ax = sns.swarmplot(x = 'gdpPerCapita', y = 'lifeExpectancy', data = data_2007)
ax.set_title("Relationship between gdp and life expectancy")
plt.show()
```

```python
# corr
data_2007.gdpPerCapita.corr(data_2007.lifeExpectancy)
```

```python
# not linear, so log10
sns.swarmplot(x = np.log10(data_2007.gdpPerCapita), y = data_2007.lifeExpectancy, data = data_2007)
plt.show()
```

```python
plt.scatter(np.log10(data_2007.gdpPerCapita), data_2007.lifeExpectancy,5)
plt.xlabel("log10 gdp per capita ($)")
plt.ylabel("life expectancy")
plt.show()
```

```python
# corr
np.log10(data_2007.gdpPerCapita).corr(data_2007.lifeExpectancy)
```

```python
# for loop
year_sorted = sorted(set(data.year))
```

```python
for given_year in year_sorted:
    data_year = data[data.year == given_year]
    plt.scatter(data_year.gdpPerCapita, data_year.lifeExpectancy,5)
    plt.xlim(0, 60000) # ignore outliers
    plt.ylim(25, 85)
    plt.title("given year")
    # plt.show()
    plt.savefig(str(given_year), dpi = 200) # dots per inch
    plt.clf # clear the current plot
```

```python
# hue=year
sns.swarmplot(x = 'gdpPerCapita', y = 'lifeExpectancy',hue = 'year', data = data)
plt.show()
```
